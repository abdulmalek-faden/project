## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Task (2) Dataset.csv')
df.head()
```

```python
df.info()
```

### Identify and analyze patterns in the data.

```python
df.dropna(subset=['TravelTimeHistoric'],inplace=True)
df.dropna(subset=['TravelTimeLive'],inplace=True)
df.isnull().sum()
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], bins=20, kde=True)
plt.title('Histogram of  Traffic Index Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Datetime'], bins=20, kde=True)
plt.title('Histogram of  Datetime')
plt.xlabel('Datetime')
plt.ylabel('Frequency')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(11,8))
df.plot.scatter(x='TrafficIndexLive', y='Weather_Condition')
plt.title('the relationship between Traffic_Index and Weather_Condition')
plt.xlabel('Traffic_Index')
plt.ylabel('Weather_Condition')
plt.show()

#the Weather_Condition not found in data set
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(11,8))
df.plot.hexbin( x='TrafficIndexLive', y='JamsLength', gridsize=20)
plt.title('the density of data points for TrafficIndexLive and JamsLength ')
plt.xlabel('TrafficIndexLive')
plt.ylabel('JamsLength')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
plt.figure(figsize=(11, 8))
sns.histplot(df['City'], kde=True, bins=30, color='red')
plt.title('histogram of city')
plt.xlabel('city')
plt.ylabel('frequency')
plt.show()

plt.figure(figsize=(12, 15))
sns.barplot(data=df, x='City', y='Datetime', palette='muted')
plt.title('Average vistor to the city by date')
plt.xlabel('city')
plt.ylabel('Datetime')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(11, 8))
df.plot.scatter(x='Datetime', y='TrafficIndexLive')
plt.title('the relationship between Datetime and TrafficIndexLive')
plt.xlabel('Datetime')
plt.ylabel('TrafficIndexLive')
plt.xticks(rotation=90)
plt.show()
```
